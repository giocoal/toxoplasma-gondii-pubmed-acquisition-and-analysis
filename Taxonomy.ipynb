```python
def taxonomy(id_, retmode = 'xml'):
   
    handle = Entrez.efetch(db="taxonomy", id=id_, retmode='xml')
    record = Entrez.read(handle)
    handle.close()
    data = record[0]
    dizionario = {}
    lista = []
    
    dizionario['ScientificName'] = data['ScientificName'] 
    for i in range(0, len(data['LineageEx'])):
      if data['LineageEx'][i]['Rank'] == 'clade':
        lista.append(data['LineageEx'][i]['ScientificName'])
        dizionario['clade'] = lista
      else:
        dizionario[data['LineageEx'][i]['Rank']] = data['LineageEx'][i]['ScientificName']
    del dizionario['no rank']
      
    return(dizionario)

tassonomia = taxonomy(id_= 5811)
tassonomia
```
